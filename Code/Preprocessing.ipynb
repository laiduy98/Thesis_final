```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


EEGsignal = pd.read_csv('5_1.tdt', sep = '\t')
```

```python
EEGsignal['O2-AVE'].plot()
plt.show()
```

```python
from scipy import stats

zscore = np.abs(stats.zscore(EEGsignal))
```

```python
zscore
```

```python
EEGsignal_cleaned = EEGsignal[(zscore < 3).all(axis=1)]
```

```python
EEGsignal
```

```python
EEGsignal_cleaned = EEGsignal_cleaned[['FP1-AVE', 'FP2-AVE', 'F3-AVE', 'F4-AVE', 'F7-AVE', 'F8-AVE', 'Fz-AVE', 'C3-AVE', 'C4-AVE', 'Cz-AVE', 'T3-AVE', 'T4-AVE', 'T5-AVE', 'T6-AVE', 'P3-AVE', 'P4-AVE', 'Pz-AVE', 'O1-AVE', 'O2-AVE']]
```

```python
EEGsignal_cleaned
```

```python
EEGsignal_cleaned.plot(subplots = True)
plt.show()
```

```python
EEGsignal_cleaned['O2-AVE'].plot()
plt.show()
```

```python
EEGsignal_cleaned.drop(EEGsignal_cleaned.tail(21).index, inplace =True)
```

```python
EEGsignal_cleaned_stacked = pd.melt(EEGsignal_cleaned)
```

```python
EEGsignal_cleaned_stacked
```

```python
EEGsignal_cleaned_stacked['variable'] = EEGsignal_cleaned_stacked['variable'].replace(['FP1-AVE', 'FP2-AVE', 'F3-AVE', 'F4-AVE','F7-AVE', 'F8-AVE','Fz-AVE'], 0)
EEGsignal_cleaned_stacked['variable'] = EEGsignal_cleaned_stacked['variable'].replace(['C3-AVE', 'C4-AVE','Cz-AVE'], 1)
EEGsignal_cleaned_stacked['variable'] = EEGsignal_cleaned_stacked['variable'].replace(['T3-AVE', 'T4-AVE', 'T5-AVE', 'T6-AVE'], 2)
EEGsignal_cleaned_stacked['variable'] = EEGsignal_cleaned_stacked['variable'].replace(['P3-AVE', 'P4-AVE', 'O1-AVE', 'O2-AVE', 'Pz-AVE'], 3)
```
